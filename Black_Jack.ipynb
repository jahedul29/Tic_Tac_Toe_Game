## Step 1
definition of Card() class

```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two','Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'King', 'Queen', 'Jack', 'Ace')
values = {'Two':2,'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'King':10, 'Queen':10, 'Jack':10, 'Ace':11}

playing = True

class Card():
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.suit + ' : ' + self.rank
```

### Step 2
Defining Deck() class

```python
class Deck():
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit, rank))
    
    def __str__(self):
        deck_collection = ''
        for card in self.deck:
            deck_collection = deck_collection + str(card) +'\n\n'
        return deck_collection
    
    def shuffle(self):
        random.shuffle(self.deck)
    
    def deal(self):
        single_card = self.deck.pop()
        return single_card
    
```

#### Step 4
Creating a Hand Class

```python
class Hand():
    def __init__(self):
        self.card = []
        self.value = 0
        self.ace = 0
        self.real_value = 0
        
    def add_card(self, card):
        self.card.append(card)
        self.value += values[card.rank]
        self.real_value = self.value
        if card.rank == 'Ace':
            self.ace += 1
        
    def adjust_for_ace(self):
        while self.value > 21 and self.ace:
            self.value -=10
            self.ace -= 1
```

#### Step 5
Creating Chips class

```python
class Chips():
    def __init__(self, total = 100):
        self.total = total
        self.bet_money = 0
        
    def win_bet(self):
        self.total += self.bet_money
        
    def lose_bet(self):
            self.total -= self.bet_money
        
        
```

## Different Function definition

```python
def take_bet(chips):
    while True:
        try:
            x = int(input("Enter bet amount : "))
        except:
            print("Enter a valid integer")
        else:
            if x > chips.total:
                print("Not Enough Balance\n\n")
            else:
                chips.bet_money = x
                break
```

```python
def hit(deck, hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()

    
```

```python
def hit_or_stand(deck, hand):
    global playing
    while True:
        decision = input("Hit or Stand? Enter 'h' or 's' : ")
        print(decision)
        if decision.lower() == 'h':
            hit(deck,hand)
            break
        elif decision.lower() == 's':
            print("\nPlayer Stands!! It's dealers turn\n\n")
            playing = False
            break
        else:
            print("Enter only 'h' or 's'")
```

```python
def player_busts(chips):
    print("\nPlayer busts\n\nDealer wins\n\n")
    chips.lose_bet()
    
def player_wins(chips):
    print("\nPlayer Wins\n\n")
    chips.win_bet()
    
def dealer_busts(chips):
    print("\nDealer busts\n\nPlayer Wins\n\n")
    chips.win_bet()
    
def dealer_wins(chips):
    print("\nDealer wins\n\n")
    chips.lose_bet()
    
def push():
    print("\nPush!! No one wins!\n\n")
```

```python
def show_some(player, dealer):
    print("DEALER'S HAND : \n")
    print("One card hidden!!\n")
    print(dealer.card[1])
    print("Total value of Dealer: {}\n\n".format(values[dealer.card[1].rank]))
    print("\n")
    
    print("PLAYER'S HAND :\n")
    for card in player.card:
        print(card)
    print("Total value of Player: {}\n\n".format(player.value))
    
def show_all(player, dealer):
    print("DEALER'S HANE : \n")
    for card in dealer.card:
        print(card)
    print("Total value : {}\n\n".format(dealer.value))    
    print("\n")
    
    print("PLAYER'S HAND :\n")
    for card in player.card:
        print(card)
    print("Total value : {}\n\n".format(player.value))
    
```

```python
def replay():
    
    decision = input('Do you want to replay the gamy ?? Y or N?')
    while decision.lower() != 'y' and decision.lower() != 'n' :
        print('Enter valid input !!\n')
        decision = input('Do you want to replay the gamy ?? Y or N? : ')
        
    if decision.lower() == 'y':
        return True
    else:
        return False
```
